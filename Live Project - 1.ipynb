```python
X = new_df.drop('Order Priority',axis=1)
y = new_df['Order Priority']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets,metrics
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
```

##Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
# split data into training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=1,test_size=0.15,shuffle=True)
# create object for logistic regression
logistic_regression = LogisticRegression(max_iter=1000)
#train the model
logistic_regression.fit(x_train, y_train)
#predict values
y_pred = logistic_regression.predict(x_test)
#check model accuracy
lr_accuracy = metrics.accuracy_score(y_test, y_pred)
lr_accuracy_percentage = 100 * lr_accuracy
print("Accuracy percentage : ",lr_accuracy_percentage)
mse=metrics.mean_squared_error(y_test, y_pred)
mse_per = 100 * mse
print("Mean squared error : ",mse_per)
print("Confusion matrix : \n",confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

##SVM

```python
from sklearn import svm
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15,random_state=1) 
model = SVC(kernel="poly",random_state=1,C=0.025)
model.fit(X_train, y_train)
model.score(X_test, y_test)
y_pred = model.predict(X_test)
svm_accuracy = metrics.accuracy_score(y_test, y_pred)
svm_accuracy_percentage = 100 * svm_accuracy
print("Accuracy percentage : ",svm_accuracy_percentage)
mse=metrics.mean_squared_error(y_test, y_pred)
mse_per = 100 * mse
print("Mean squared error : ",mse_per)
print("Confusion matrix : \n",confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

##KNN


```python
from sklearn.neighbors import KNeighborsClassifier
error_rate = []
for i in range(1,30):
  knn = KNeighborsClassifier(i)
  knn.fit(X_train,y_train)
  pred=knn.predict(X_test)
  error_rate.append(np.mean(pred!=y_test))
```

```python
plt.plot(range(1,30),error_rate,color='#b5041e')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15,random_state=0) 
knn=KNeighborsClassifier(n_neighbors=28)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
knn_accuracy = metrics.accuracy_score(y_test, y_pred)
knn_accuracy_percentage = 100 * knn_accuracy
print("Accuracy percentage : ", knn_accuracy_percentage)
mse=metrics.mean_squared_error(y_test, y_pred)
mse_per = 100 * mse
print("Mean squared error : ",mse_per)
print("Confusion matrix : \n",confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

##Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15,random_state=10,shuffle=True) 
dtree=DecisionTreeClassifier(max_depth=11,random_state=0,max_features=None)
dtree.fit(X_train,y_train)
y_pred=dtree.predict(X_test)
dtree_accuracy = metrics.accuracy_score(y_test, y_pred)
dtree_accuracy_percentage = 100 * dtree_accuracy
print("Accuracy percentage : ", dtree_accuracy_percentage)
mse=metrics.mean_squared_error(y_test, y_pred)
mse_per = 100 * mse
print("Mean squared error : ",mse_per)
print("Confusion matrix : \n",confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

##Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.15, random_state=10)
rf = RandomForestClassifier(n_estimators=1500, random_state=0,max_depth=100)
rf.fit(train_x,train_y)
y_pred = rf.predict(test_x)
rf_accuracy = metrics.accuracy_score(test_y, y_pred)
rf_accuracy_percentage = 100 * rf_accuracy
print("Accuracy percentage : ", rf_accuracy_percentage)
mse=metrics.mean_squared_error(test_y, y_pred)
mse_per = 100 * mse
print("Mean squared error : ",mse_per)
print("Confusion matrix : \n",confusion_matrix(test_y, y_pred))
print(classification_report(test_y,y_pred))
```

##Gradient Boosting
